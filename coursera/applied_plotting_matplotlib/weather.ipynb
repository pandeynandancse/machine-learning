```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss

```

```python
import seaborn as sns

%matplotlib notebook
```

```python
plt.style.available

plt.style.use('seaborn-colorblind')
df = pd.read_csv('Min_Max_Seasonal_IMD_2017.csv')
df
```

```python
a= df['YEAR']
a = a[98:]
li=[]
for i in range(len(a)):
    x = a.loc[98+i]
    li.append(x)
    
plt.plot(li,df['ANNUAL - MIN'][98:],label = 'Annual Minimum Temprature of India')
plt.plot(li,df['ANNUAL - MAX'][98:],label='Annual Maximum Temprature of India')
plt.xlabel("Year")
plt.ylabel("Temprature")
plt.xticks(a,rotation=45)
plt.legend(frameon = True).get_frame().set_edgecolor('b')
plt.title("Annual Min. and Max. temprature of India(1999-2017)")
plt.fill_between(li,df['ANNUAL - MIN'][98:],df['ANNUAL - MAX'][98:],facecolor='green',alpha='.5')
```
